## _Data_

### To analyze the best place to open a bar in Crete, Greece , we need the following data sources: 

1. Data regarding the number of neighbourhoods in Crete , Greece 
2. Longitude and latitude of each neighbourhood in Crete , Greece 
3. Use tourism attraction sites combines with neighborhoods to determine more favourable neighborhoods 
4. Venue data particularly bars , but also focusing on nearby venues that could increase clientele such as hotels, restaurants and more 

### Sources of the data are:

1. Cities in Crete , Greece will be obtained from Wikipedia using the link [Greece] https://en.wikipedia.org/wiki/List_of_cities_and_towns_in_Greece
   - This link will provide us with all Greek regions and towns. 
   - We will then have to filter for regions in Crete. 
   - We will use BeautifulSoup to convert the data into python-workable format and creat a data frame with neighbourhoods and locations specific such as postal code and more. 

2. Download latitude and longitude from the Greek site : https://simplemaps.com/data/gr-cities ( csv format uploaded on notebook)

3. Foursquare used to find venues near specifically for the main cities in Crete ( using bars, cafe , hotels and restaurants)

4. Use the airport list in Crete to determine most preferrable location to open a bar [Airports] https://en.wikipedia.org/wiki/List_of_airports_in_Crete

5. Use k-means and one hit encoding to analyse data and figure out which place is better to open a bar in Crete.
